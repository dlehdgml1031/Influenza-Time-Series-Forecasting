```python
import pandas as pd

news1 = pd.read_csv('filtered_news_title_1.csv')
news2 = pd.read_csv('filtered_news_title_2.csv')

# 병합

news3 = pd.concat([news1, news2], axis = 0)
news3
```

```python
news = news3.reset_index()
news
```

```python
news = news.sort_values(by='Date')
news_cleaned = news
```

```python
news_cleaned['Title'] = news_cleaned['Title'].astype(str)
news_cleaned
```

```python
import re

def remove_special_chars(text):
    return re.sub(r'[()\[\]{}\-:;]', '', text)

news_cleaned['Title'] = news_cleaned['Title'].apply(remove_special_chars)
```

```python
# 영어만, 태그가 <title> 인 것만 살리기, 태그가 없으면 가만히 냅두기

def clean_html_except_title(text):
    cleaned = re.sub(r'<(?!title\b)[^>]*>.*?</[^>]+>', '', text)
    final_cleaned = re.sub(r'</?title>', '', cleaned)
    return final_cleaned

news_cleaned['Cleaned_Title'] = news_cleaned['Title'].apply(clean_html_except_title)

def contains_only_english(text):
    return bool(re.match(r'^[a-zA-Z0-9\s.,!?\'\"()\-]+$', text))

news_cleaned = news_cleaned[news_cleaned['Cleaned_Title'].apply(contains_only_english)]
```

```python
stop_words = set(['[',']', '{', '}', '--', ':', ';', '<', '>'])

def process_text(text):
    words = text.lower().split()
    words = [word for word in words if word not in stop_words]
    return ' '.join(words)

news_cleaned['Title'] = news_cleaned['Title'].apply(process_text)
```

```python
news_cleaned = news_cleaned.drop_duplicates(subset='Title', keep='first')
news_cleaned
```

```python
news_cleaned = news_cleaned[['Title', 'Date']]
news_cleaned
```

```python
news_input = news_cleaned
news_input
```

```python
from datetime import datetime, timedelta

def parse_date(date_str):
    for fmt in ("%Y%m%d", "%Y-%m-%d"):
        try:
            return datetime.strptime(date_str, fmt)
        except ValueError:
            pass
    return None  # 만약 포맷에 맞지 않으면 None 반환

# PTNAPLC_YMD 열 변환
news_input['Date'] = news_input['Date'].apply(parse_date)
```

```python
news_input
```

```python
df_for_input = news_input
```

```python
def convert_to_datetime(row):
    try:
        return pd.to_datetime(row['Date'], format='%Y-%m-%d', errors='raise')
    except Exception as e:
        return None

df_for_input['Date'] = df_for_input.apply(convert_to_datetime, axis=1)

df_for_input = df_for_input.dropna(subset=['Date'])

df_for_input
```

```python
import numpy as np
from datetime import datetime, timedelta

def to_weekly_date_custom(date):
    adjusted_date = date - timedelta(days=1)  
    monday = adjusted_date - timedelta(days=adjusted_date.weekday())  
    return monday + timedelta(days=1)  


def to_weekly_date_custom_with_error_handling(row):
    try:
        return to_weekly_date_custom(row['Date'])
    except Exception as e:
        return None

df_for_input['Weekly_Date_Custom'] = df_for_input.apply(to_weekly_date_custom_with_error_handling, axis=1)

df_for_input = df_for_input.dropna(subset=['Weekly_Date_Custom'])

df_for_input
```

```python
df_for_input = df_for_input.sort_values(by='Weekly_Date_Custom')
df_for_input
```

```python
title_counts = df_for_input.groupby('Weekly_Date_Custom')['Title'].count().reset_index(name='Title_Count')
title_counts.to_csv('news_count.csv')
```

```python
df_for_input['News_Index'] = df_for_input.groupby('Weekly_Date_Custom').cumcount() + 1
df_for_input['Updated_Title'] = '[News ' + df_for_input['News_Index'].astype(str) + '] ' + df_for_input['Title']

# Now we will concatenate all titles within each 'Weekly_Date_Custom' group
grouped_df = df_for_input.groupby('Weekly_Date_Custom')['Updated_Title'].apply(' '.join).reset_index()
grouped_df
```
